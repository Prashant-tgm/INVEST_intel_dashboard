# 04 - Risk Assessment Engine

Goal: assess historical stock and portfolio risk with volatility, Sharpe, Sortino, drawdown, VaR, and expected shortfall.

```python
from pathlib import Path
import sys

ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.append(str(ROOT / 'src'))

import plotly.express as px

from invest_intel.data import build_returns_matrix, load_prices
from invest_intel.risk import max_drawdown, portfolio_metrics, portfolio_return_series, risk_summary
```

```python
prices = load_prices()
returns = build_returns_matrix(prices[prices['date'] >= '2015-01-01'], min_obs=252)
risk = risk_summary(returns)
risk.head(20)
```

```python
px.scatter(
    risk,
    x='annualized_volatility',
    y='annualized_return',
    color='sharpe_ratio',
    hover_name='symbol',
    title='Risk-return map by stock',
)
```

```python
px.bar(risk.sort_values('max_drawdown').head(20), x='symbol', y='max_drawdown', title='Largest historical drawdowns')
```

```python
equal_weight_symbols = risk.head(10)['symbol']
weights = {symbol: 1 / len(equal_weight_symbols) for symbol in equal_weight_symbols}
portfolio_metrics(returns, weights)
```
