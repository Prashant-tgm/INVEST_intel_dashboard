# 03 - Portfolio Construction Engine

Goal: recommend transparent portfolios for conservative, balanced, and aggressive investor profiles.

```python
from pathlib import Path
import sys

ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.append(str(ROOT / 'src'))

import pandas as pd
import plotly.express as px

from invest_intel.data import build_returns_matrix, load_prices
from invest_intel.portfolio import construct_profile_portfolio, random_portfolios
```

```python
prices = load_prices()
recent_prices = prices[prices['date'] >= '2015-01-01']
returns = build_returns_matrix(recent_prices, min_obs=252).dropna(how='all')
returns.shape
```

```python
profile_outputs = {}
for profile in ['conservative', 'balanced', 'aggressive']:
    weights, metrics, selected = construct_profile_portfolio(returns, profile=profile)
    profile_outputs[profile] = {'weights': weights, 'metrics': metrics, 'selected': selected}
    print(profile, metrics)
    display(selected[['symbol', 'weight', 'expected_return', 'annualized_volatility', 'sharpe_score']])
```

```python
balanced = profile_outputs['balanced']['selected']
px.bar(balanced, x='symbol', y='weight', title='Balanced profile allocation')
```

```python
frontier = random_portfolios(returns.dropna(), n_portfolios=1000)
px.scatter(frontier, x='annualized_volatility', y='expected_return', color='sharpe_ratio', title='Random portfolio frontier')
```
